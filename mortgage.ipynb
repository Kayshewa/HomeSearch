```python
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import http.client
import json
import pandas as pd
import plotly.io as pio
import os
from datetime import datetime
```

```python
# Connect to the Zillow API
conn = http.client.HTTPSConnection("zillow-com1.p.rapidapi.com")

headers = {
    'x-rapidapi-key': "287c0c69cbmsh41c226a35114322p1022a8jsn65498208884c",
    'x-rapidapi-host': "zillow-com1.p.rapidapi.com"
}

conn.request("GET", "/mortgageRates?program=Fixed30Year&state=TX&refinance=false&loanType=Conventional&loanAmount=Conforming&loanToValue=Normal&creditScore=VeryHigh&duration=4000", headers=headers)

# Get the response
res = conn.getresponse()
data = res.read()

# Decode the byte data to a string and then parse it as JSON
mortgage_data = json.loads(data.decode("utf-8"))

# Access the 'samples' list inside the 'Fixed30Year' program
samples = mortgage_data['rates']['Fixed30Year']['samples']
```

```python
# Extract 'time' and 'rate' fields from each entry in samples
times = [sample['time'] for sample in samples]
rates_values = [sample['rate'] for sample in samples]

# Create a pandas DataFrame
df_rates = pd.DataFrame({'Time': times, 'Rate': rates_values})

# Display the DataFrame
print(df_rates)
```

```python
#TURTLEWEED
# Make the API request
conn.request("GET", "/zestimateHistory?zpid=338894010", headers=headers)

# Get the response
res = conn.getresponse()
data = res.read()

# Decode the byte data to a string and then parse it as JSON
zestimate_data = json.loads(data.decode("utf-8"))

# Check if the data contains the expected structure
if isinstance(zestimate_data, list):
    # Extract 't' and 'v' fields
    df_zestimate_turtleweed = pd.DataFrame({'t': [entry['t'] for entry in zestimate_data],
                                            'v': [entry['v'] for entry in zestimate_data]})
    
    # Convert timestamp to datetime format
    df_zestimate_turtleweed['t'] = pd.to_datetime(df_zestimate_turtleweed['t'], unit='s')
    
    # Sort the data by timestamp
    df_zestimate_turtleweed = df_zestimate_turtleweed.sort_values('t')
    
    # Format the Zestimate values as USD for display, but keep original values for plotting
    df_zestimate_turtleweed['v_display'] = df_zestimate_turtleweed['v'].apply(lambda x: "${:,.2f}".format(x))
    
    # Display the DataFrame
    print(df_zestimate_turtleweed)
else:
    print("Unexpected data format:", zestimate_data)
```

```python
#BERING
# Make the API request
conn.request("GET", "/zestimateHistory?zpid=28370490", headers=headers)

# Get the response
res = conn.getresponse()
data = res.read()

# Decode the byte data to a string and then parse it as JSON
zestimate_data = json.loads(data.decode("utf-8"))

# Check if the data contains the expected structure
if isinstance(zestimate_data, list):
    # Extract 't' and 'v' fields
    df_zestimate_bering = pd.DataFrame({'t': [entry['t'] for entry in zestimate_data],
                                            'v': [entry['v'] for entry in zestimate_data]})
    
    # Convert timestamp to datetime format
    df_zestimate_bering['t'] = pd.to_datetime(df_zestimate_bering['t'], unit='s')
    
    # Sort the data by timestamp
    df_zestimate_bering = df_zestimate_bering.sort_values('t')
    
    # Format the Zestimate values as USD for display, but keep original values for plotting
    df_zestimate_bering['v_display'] = df_zestimate_bering['v'].apply(lambda x: "${:,.2f}".format(x))
    
    # Display the DataFrame
    print(df_zestimate_bering)
else:
    print("Unexpected data format:", zestimate_data)
```
